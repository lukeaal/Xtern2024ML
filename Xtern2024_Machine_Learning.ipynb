# **Build and Train Model**
### (Outline) Model Selection, Training & Testing, and Data Prep
The type of machine learning problem we are trying to perform is classifying the features into a small discrete numbers of types of orders. This is a textbook classification problem.

Our data is purely catagoricaly, no continuous values here. Our data is uniformly distributed on the targets but biased in the feautures.

The model resulting from this project should be able to run quickly on a phone or by API call. The response time should be quick as to not slow down the users expreience.

The model should be pre-trained, meaning that once a user uses it, predictions are fast.

Our data set only has around 5000 points and 5 features, deep underlying connections between features are unlikely in such dat sets.

Training this model should be cheap but balanced with a focus on accuracy. As to save moeny training and giving out too many discounts.

With all of these things considered; a simple, quick to train, pre-train-able model, with few prior assuptions about the features relationships would be a **Decision Tree** with some fine tuned pruning.

I wil traing the model using SKlearn's `Tree` Class. I will test the model with SKlearn's `TESTING TESING TESTING`



```python
# import the tree from SKlearn to use as the model
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import OneHotEncoder
```

```python
year_mapping = {'Year 2': 2, 'Year 3': 3, 'Year 1': 1, 'Year 4': 4}
data['Year'] = data['Year'].map(year_mapping)

X = data[['Major', 'Time', 'Year', 'University']]
Y = data['Order']

# One-hot encode categorical features
encoder = OneHotEncoder(sparse=False, drop='first')
X_encoded = encoder.fit_transform(X[['Major', 'University']])
feature_names = encoder.get_feature_names_out(['Major', 'University'])
X_encoded = pd.DataFrame(X_encoded, columns=feature_names)
X_encoded[['Time', 'Year']] = X[['Time', 'Year']]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, Y, test_size=0.2, random_state=42)

# Create a decision tree classifier and train
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)

# Make predictions on the test set
predictions = clf.predict(X_test)

# Find the accuracy of predictions
accuracy = clf.score(X_test, y_test)
print("Accuracy:", accuracy)

# Make predictions on the test set
predictions = clf.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)

# Generate a classification report
print("Classification Report:")
print(classification_report(y_test, predictions))
```
